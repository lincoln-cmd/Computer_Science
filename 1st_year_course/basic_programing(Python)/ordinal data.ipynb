```python
# 리스트

a = [1, 2, 3, 4]
print(a)
print(a[0])
print(a[1:3]) # 리스트의 슬라이싱은 서브 리스트를 반환

b = [1,2,3,[4,5,6]] # 다중 리스트
print(b[3][0])
print(b[3][1:3])
```

```python
# 리스트 관련 함수

# 요소 추가
a = [1,2,3]
a.append(5)
print('a: ', a)

# 정렬
b = [5,2,3,4,1]
b.sort()
print('b: ', b)

# 위치 반환
c = [1,2,3,4]
print('c: ', c.index(3))

# 삭제
d = [1,2,3,4]
del d[3]
print('d: ', d)
```

```python
# sort와 sorted

a = [5,4,3,2,1]
a.sort()
print('a: ', a)
b = [4,3,2,1]
sorted(b)
print('b: ', b)
```

```python
# 튜플

a = (1,2,3)
b = (1)
c = (1,) # 튜플의 요소가 1개일 때는 반드시 ','를 붙여야 한다.

print(type(a))
print(type(b))
print(type(c))
```

```python
# set

a = {1,2,3}
b = set([4,5,2,3,1])
c = 'Hello World'

print(a)
print(b)
print(set(c)) # set은 중복을 허용하지 않고, 순서도 없다.
```

```python
# set의 연산자들

a = {1,2,3,4}
b = {3,4,5,6}
c = {5,6,7,8}

a.add(5)
print(a)

c.clear()
print(c)

# 합집합
print('합집합:', a.union(b))

# 교집합
print('교집합:', a.intersection(b))

# 차집합:
print('차집합1:', a.difference(b))
print('차집합2:', b.difference(a))
```

```python
# list, tuple, set의 변환

a,b,c = [1,2,3], (1,2,3), {1,2,3}
print('a:', type(a))
print('b:', type(b))
print('c:', type(c))

a,b,c = tuple(a), set(b), list(c)
print('a:', type(a))
print('b:', type(b))
print('c:', type(c))
```
