```python
# 매핑형(딕셔너리)

a = {'1':'홍길동', '2':'이순신', '3':'무지개', 4:'심봉사'}
print(a)

print(a['1'])
print(a[4])
```

```python
# 딕셔너리 생성

b = dict()
b['1'] = '홍길동'
b['t'] = 'tuple'

print(b)
```

```python
# 매핑형 자료의 주요 함수들

a = {'1':'홍길동', '2':'이순신', '3':'무지개', 4:'심봉사'}

print(a.keys())

print(a.values())

del a['1']
print(a)

# a['1'] # 해당 key 값이 없을 경우 에러 발생
a.get('1') # 해당 key 값이 없을 경우 아무것도 반환하지 않음
```
