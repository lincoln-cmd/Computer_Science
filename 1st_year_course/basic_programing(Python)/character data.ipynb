```python
# escape 문자

print('Hello \n World')
print('Hello \t World')
print('Hello \' World')
```

```python
# 문자열 더하기, 곱하기

print("Hello" + "World")
print("Hello World" * 3)
```

```python
# 인덱싱과 슬라이싱

A = "Life is short, you need python"
print(A[0])
print(A[-1])
print(A[-2])
print(A[5:]) # 5번째 문자부터 끝까지
print(A[:5]) # 처음부터 5번째 문자 직전까지
print(A[1:4]) # 1번째 문자부터 4번째 문자 직전까지 (1번째 문자 ~ 3번째 문자)
print(A[2:-2])
```

```python
# 주요 문자열 함수

print(len('Life is short'))

print('Life is short'.startswith('Life'))
print('Life is short'.startswith('Wife'))

print('Life is short'.endswith('short'))
print('Life is short'.endswith('long'))

print('Life is short'.find('is'))
print('Life is short'.find('long'))

print('Life is short'.count('i'))

print('       Life is short   '.strip())
```

```python
# 문자열 포맷팅

print('현재 시간은 {}시 입니다.'.format(7))
print('현재 시간은 {0}시 {1}분 입니다.'.format(3, 10))
print('현재 시간은 %d시 %d분 입니다.'%(2, 30))
print('현재 시간은 {time}시 {minute}분 입니다.'.format(time = 12, minute = 13))
```
