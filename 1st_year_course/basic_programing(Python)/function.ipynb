```python
# 함수

def func(x):
    y = 4*x**2+2*x-4
    return y

result = func(3)
print(result)
```

```python
def test_func(x, y):
    return x - y, y - x

result3 = test_func(10, 5)
print(result3)
```

```python
a, b = test_func(10, 5)
print('a: ', a)
print('b: ', b)
```
