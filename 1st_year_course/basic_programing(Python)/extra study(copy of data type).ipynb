```python
# 추가: 자료형의 복사

# 숫자형
a = 1
b = a
print('b:', b)
print('a is equal to b:', a == b)
```

```python
b += 1
print('a:', a)
print('b:', b)
print('a is equal to b:', a == b)
```

```python
# 리스트형

a = [1, 2, 3]
b = a
print('b:', b)
print('a is equal to b:', a == b)
```

```python
b[0] = 4
print('a:', a)
print('b:', b)
print('a is equal to b:', a == b)
```

```python
# copy() 함수 활용

a = [1,2,3]
b = a.copy()
print('b:', b)
print('a is equal b:', a == b)
```

```python
b[0] = 4
print('a:', a)
print('b:', b)
print('a is equal b:', a == b)
```

```python
# deepcopy() 함수 활용

a = [1,2,[3,4]]
b = a.copy()
print('b:', b)
print('a is equal to b:', a == b)
```

```python
b[0] = 5
print('a:', a)
print('b', b)
print('a is equal to b:', a == b)
```

```python
b = a.copy()
b[2][0] = 5
print('a:', a)
print('b:', b)
print('a is equal to b:', a == b)
```

```python
import copy
a = [1,2,[3,4]]
b = copy.deepcopy(a)
print('a:', a)
print('b:', b)
print('a is equal b:', a == b)
```

```python
b[2][0] = 5
print('a:', a)
print('b:', b)
print('a is equal to b:', a == b)
```
